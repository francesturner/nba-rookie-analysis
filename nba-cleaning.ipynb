```python
#checking format name players
pattern = re.compile(r'[^a-zA-Z0-9\s\-\']')
players_with_strange_chars = data1[data1['Player'].apply(lambda x: bool(pattern.search(x)))]
print(players_with_strange_chars)
```

```python
data1['Player']=data1['Player'].str.replace('\u00A0',' ')

# Function to clean player names
def clean_player_name(name):
    # Replace specific unwanted characters
    name = name.replace('Ä', 'c')  # Example of replacing 'Ä' with 'c'
    # Remove any special characters using regex
    name = re.sub(r'[^\w\s]', '', name)  # This removes all non-word characters except spaces
    return name

# Apply the cleaning function to the 'Player' column
data1['Player'] = data1['Player'].apply(clean_player_name)

# Also strip any leading or trailing whitespace
data1['Player'] = data1['Player'].str.strip()
```

```python
print("Original DataFrame:")
print(nba22_rookie_value)

# Function to clean and convert dollar values to float
def clean_and_convert(value):
    if isinstance(value, str):  # Check if the value is a string
        value = value.replace('$', '')  # Remove dollar sign
        value = value.replace(',', '')  # Remove commas
        return value.strip()  # Trim whitespace
    return value  # Return non-string values as they are

nba22_rookie_value[['Value', 'Average Annual Value']] = nba22_rookie_value[['Value', 'Average Annual Value']].applymap(clean_and_convert)

# Convert the cleaned values to float for both columns
nba22_rookie_value[['Value', 'Average Annual Value']] = nba22_rookie_value[['Value', 'Average Annual Value']].apply(pd.to_numeric, errors='coerce')

# Display the modified DataFrame
print("\nModified DataFrame:")
print(nba22_rookie_value)
```
