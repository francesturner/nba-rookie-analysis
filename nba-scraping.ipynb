```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import lxml
```

```python
#EXTRACTING AND WEB SCRAPING
#22 NBA per game stats

url = "https://www.basketball-reference.com/leagues/NBA_2023_per_game.html"
html= requests.get(url).text
soup = BeautifulSoup (html, 'html.parser')
```

```python
tables = soup.find('table')

for table in tables:
    print(tables.text)
```

```python
for row in soup.find('tbody').find_all('tr'):
    col = row.find_all('td')
    print([c.text for c in col])  # Print all columns to see their order
```

```python
nba_pergame_stats= pd.DataFrame(columns=['Player', 'Age', 'Team', 'Games', 'Rebounds', 'Assists', 'Steals', 'Points per game'])

for row in soup.find('tbody').find_all('tr'):
    col=row.find_all('td')
    if len(col) > 0:
        Player=col[0].text
        Age=col[1].text
        Team=col[2].text
        G=col[4].text
        TRB=col[22].text
        AST=col[23].text
        STL=col[24].text
        PTS=col[28].text
        
        nba_pergame_stats=pd.concat([nba_pergame_stats, pd.DataFrame({'Player':[Player], 'Age':[Age], 'Team':[Team],'Games':[G],'Rebounds':[TRB],'Assists':[AST],'Steals':[STL],'Points per game':[PTS]})],
        ignore_index=True)

print(nba_pergame_stats)
```

```python
df=nba_pergame_stats
print(df)
nba22_pergame_stats=df
```

```python
#22 NBA drafts
url = "https://en.wikipedia.org/wiki/2022_NBA_draft#Draft_picks"
html= requests.get(url).text
soup = BeautifulSoup (html, 'html.parser')
```

```python
tables = soup.find_all('table')
print(tables)
```

```python
table = tables[3]  # Python uses 0-based indexing, so tables[1] is the second table
print(table)
```

```python
rows = table.find_all('tr')

nba22_draft_picks = pd.DataFrame(columns=['Round', 'Pick', 'Player', 'Nationality'])

for row in rows:
    col = row.find_all('td')
    
    # Check if the first column has data, even if others are missing
    if len(col) > 0:
        Rnd = col[0].text.strip() if len(col) > 0 else None
        Pick = col[1].text.strip() if len(col) > 1 else None
        Player = col[2].text.strip() if len(col) > 2 else None
        Nationality = col[4].text.strip() if len(col) > 3 else None
        
        # Append only if Rnd, Pick, or Player is not None
        if Rnd or Pick or Player:
            nba22_draft_picks = pd.concat([nba22_draft_picks, pd.DataFrame({'Round': [Rnd], 'Pick': [Pick], 'Player': [Player], 'Nationality': [Nationality]})],
                                           ignore_index=True)
        else:
            print("Row skipped:", [cell.text.strip() for cell in col])  # For debugging

print(nba22_draft_picks)
```

```python
#22 NBA rookie's value
url = "https://www.spotrac.com/nba/draft/_/year/2022/sort/pick"
html= requests.get(url).text
soup = BeautifulSoup (html, 'html.parser')
```

```python
tables = soup.find_all('table')
print(tables)
```

```python
table=tables[0]
print(table)
```

```python
rows = table.find_all('tr')

nba22_rookie_value = pd.DataFrame(columns=['Player', 'Years Contract', 'Value', 'Average Annual Value'])

for row in rows:
    col = row.find_all('td')
    
    # Check if the first column has data, even if others are missing
    if len(col) > 0:
        Player = col[4].text.strip() if len(col) > 0 else None
        yrs = col[8].text.strip() if len(col) > 0 else None
        value = col[9].text.strip() if len(col) > 2 else None
        aav = col[10].text.strip() if len(col) > 3 else None
        
        # Append only if Rnd, Pick, or Player is not None
        if Rnd or Pick or Player:
            nba22_rookie_value = pd.concat([nba22_rookie_value, pd.DataFrame({'Player': [Player], 'Years Contract': [yrs], 'Value': [value], 'Average Annual Value': [aav]})],
                                           ignore_index=True)
        else:
            print("Row skipped:", [cell.text.strip() for cell in col])  # For debugging

print(nba22_rookie_value)
```
